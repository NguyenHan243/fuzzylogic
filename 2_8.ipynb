```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
heart = ctrl.Antecedent(np.arange(40,120.1,0.1),'heart')
pressure = ctrl.Antecedent(np.arange(70,150.1,0.1),'pressure')
temp= ctrl.Antecedent(np.arange(35,40.1,0.1),'temp')
oxygen=ctrl.Antecedent(np.arange(70,100.1,0.1),'oxyen')
sugar=ctrl.Antecedent(np.arange(50,180.1,0.1),'sugar')

lieu=ctrl.Consequent(np.arange(0,10.1,0.1),'lieu')
dich=ctrl.Consequent(np.arange(0,10.1,0.1),'dich')
oxy=ctrl.Consequent(np.arange(0,10.1,0.1),'oxy')

heart['low']=fuzz.gaussmf(heart.universe,50,20)
heart['normal']=fuzz.gaussmf(heart.universe,80,20)
heart['high']=fuzz.gaussmf(heart.universe,120,30)
pressure['low']=fuzz.gaussmf(pressure.universe,80,20)
pressure['normal']=fuzz.gaussmf(pressure.universe,110,20)
pressure['high']=fuzz.gaussmf(pressure.universe,140,20)
temp['low']=fuzz.gaussmf(temp.universe,35,2)
temp['normal']=fuzz.gaussmf(temp.universe,37.5,1.5)
temp['high']=fuzz.gaussmf(temp.universe,39,2.5)
oxygen['low']=fuzz.trimf(oxygen.universe,[70,70,90])
oxygen['normal']=fuzz.trimf(oxygen.universe,[90,95,100])
sugar['low']=fuzz.trimf(sugar.universe,[50,50,70])
sugar['normal']=fuzz.trimf(sugar.universe,[70,100,140])
sugar['high']=fuzz.trapmf(sugar.universe,[140,150,180,180])
lieu['D']=fuzz.trimf(lieu.universe,[0,2.5,5])
lieu['M']=fuzz.trimf(lieu.universe,[2.5,5,7.5])
lieu['I']=fuzz.trapmf(lieu.universe,[5,7.5,10,10])

dich['R']=fuzz.trimf(dich.universe,[0,2.5,5])
dich['M']=fuzz.trimf(dich.universe,[2.5,5,7.5])
dich['I']=fuzz.trapmf(dich.universe,[5,7.5,10,10])
oxy['R']=fuzz.trimf(oxy.universe,[0,2.5,5])
oxy['M']=fuzz.trimf(oxy.universe,[2.5,5,7.5])
oxy['I']=fuzz.trapmf(oxy.universe,[5,7.5,10,10])
```

# Mục mới

```python
rule1=ctrl.Rule(heart['low']&pressure['low'],dich['I'],lieu['I'])
rule2=ctrl.Rule(heart['high']&pressure['high'],lieu['D'])
rule3=ctrl.Rule(heart['normal']&pressure['normal'],lieu['M'])
rule4=ctrl.Rule(temp['high']&oxygen['low'],oxy['I'])
rule5=ctrl.Rule(temp['normal']&oxygen['normal'],dich['M'],lieu['M'])
rule6=ctrl.Rule(temp['normal']&oxygen['normal'],oxy['M'])
rule7=ctrl.Rule(temp['low']&oxygen['low'],oxy['I'])
rule8=ctrl.Rule(sugar['high'],lieu['I'])
rule9=ctrl.Rule(sugar['low'],lieu['D'])
rule10=ctrl.Rule(sugar['normal'],lieu['M'])
rule11=ctrl.Rule(pressure['high']&sugar['high']&heart['high'],lieu['D'])
rule12=ctrl.Rule(pressure['low']&oxygen['low']&heart['low'],lieu['I'],dich['I'])
rule13=ctrl.Rule(pressure['low']&oxygen['low']&heart['low'],oxy['I'])
rule14=ctrl.Rule(temp['high']&pressure['high']&oxygen['normal'],lieu['I'])
```
